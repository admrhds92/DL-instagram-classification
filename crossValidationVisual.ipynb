```python
import numpy as np
import matplotlib.pyplot as plt
```

<h1>Loading in vgg16, resNet50V2, and efficientNetB2 data</h1>

```python
vgg16 = np.loadtxt('vgg16/vgg16results.txt')
resNet50v2 = np.loadtxt('resNet50/resNet50results.txt')
effNet = np.loadtxt('effNet/effNetresults.txt')
```

<h1>Plot Results</h1>

```python
plt.plot(np.arange(0,vgg16.shape[1]), np.mean(vgg16,0),label="VGG16")
plt.fill_between(np.arange(0,vgg16.shape[1]),
 np.mean(vgg16,0)-(1.96*np.std(vgg16,0)/np.sqrt(vgg16.shape[0])),
 np.mean(vgg16,0)+(1.96*np.std(vgg16,0)/np.sqrt(vgg16.shape[0])),
 alpha=0.5)

plt.plot(np.arange(0,resNet50v2.shape[1]), np.mean(resNet50v2,0),label="ResNet50")
plt.fill_between(np.arange(0,resNet50v2.shape[1]),
 np.mean(resNet50v2,0)-(1.96*np.std(resNet50v2,0)/np.sqrt(resNet50v2.shape[0])),
 np.mean(resNet50v2,0)+(1.96*np.std(resNet50v2,0)/np.sqrt(resNet50v2.shape[0])),
 alpha=0.5)

plt.plot(np.arange(0,effNet.shape[1]), np.mean(effNet,0),label="Efficient Net")
plt.fill_between(np.arange(0,effNet.shape[1]),
 np.mean(effNet,0)-(1.96*np.std(effNet,0)/np.sqrt(effNet.shape[0])),
 np.mean(effNet,0)+(1.96*np.std(effNet,0)/np.sqrt(effNet.shape[0])),
 alpha=0.5)



plt.ylim([0.0,1.0])
plt.ylabel('Validation Accuracy (P)')
plt.xlabel('Epoch')
plt.legend(loc='lower right')
plt.title('Instagram Screenshot Classification - 3 Models')
plt.savefig('foo.png')
plt.show()

```
